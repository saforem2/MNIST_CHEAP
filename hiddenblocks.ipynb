```python
#hierarchical version
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
def sig(x):
    return 1/(1+np.exp(-x))
training_data_file =open(os.path.expanduser("~/deepLdata/MNIST/mnist_train.csv"))
training_data_list = training_data_file.readlines()
training_data_file.close()

test_data_file =open(os.path.expanduser("~/deepLdata/MNIST/mnist_test.csv"))
test_data_list = test_data_file.readlines()
test_data_file.close()

inodes=784
hnodes=196
onodes=10
lr1=0.5
lr2=0.1
sigih=pow(hnodes, -0.5)
sigho=pow(onodes, -0.5)



wih = np.random.normal(0.0,sigih,(hnodes,4))
who = np.random.normal(0.0,sigho,(onodes,hnodes))
hidlin = np.zeros(hnodes)
d1=np.zeros((hnodes,4))

nepochs=15
for epoch in range(nepochs):
    print(epoch)

    for aline in training_data_list:
        splitline=aline.split(',')
        correct_label = int(splitline[0])
        inputs = np.asfarray(splitline[1:])/255.0
        #hidlin = np.dot(wih, inputs)

        for k1 in range(1,15):
            for k2 in range(1,15):
                k=k1+14*(k2-1)
                jj=2*k1+56*k2-57
                hidlin[k-1]=(wih[k-1,0]*inputs[jj-1]+wih[k-1,1]*inputs[jj]
                             +wih[k-1,2]*inputs[jj+27]+wih[k-1,3]*inputs[jj+28])
        hid=sig(hidlin)
        ylin=np.dot(who,hid)
        yfin=sig(ylin)
        targets = np.zeros(onodes) + 0.0
        targets[int(correct_label)] = 1.0
        temperror1=(yfin-targets)*(yfin-yfin*yfin)
        temperror2=np.dot(temperror1,who)
        d2=np.outer(temperror1,hid)
        #d1=np.outer(temperror2*(hid-hid*hid),inputs)
        for k1 in range(1,15):
            for k2 in range(1,15):
                k=k1+14*(k2-1)
                jj=2*k1+56*k2-57
                kpart=temperror2[k-1]*(hid[k-1]-hid[k-1]*hid[k-1])
                d1[k-1,0]=kpart*inputs[jj-1]
                d1[k-1,1]=kpart*inputs[jj]
                d1[k-1,2]=kpart*inputs[jj+27]
                d1[k-1,3]=kpart*inputs[jj+28]
                pass

        wih+=-lr1*d1
        who+=-lr2*d2
        pass
    attempts=0
    success=0
    for  aline in test_data_list:
        splitline=aline.split(',')
        correct_label = int(splitline[0])
        inputs = np.asfarray(splitline[1:])/255.0
        #hidlin = np.dot(wih, inputs)
        for k1 in range(1,15):
            for k2 in range(1,15):
                k=k1+14*(k2-1)
                jj=2*k1+56*k2-57
                hidlin[k-1]=(wih[k-1,0]*inputs[jj-1]+wih[k-1,1]*inputs[jj]
                             +wih[k-1,2]*inputs[jj+27]+wih[k-1,3]*inputs[jj+28])
        hid=sig(hidlin)
        ylin=np.dot(who,hid)
        yfin=sig(ylin)
        attempts+=1
        label = np.argmax(yfin)
        if (label == correct_label):
            success+=1
        pass
    performance=(1.0*success)/attempts
    print(performance)
    pass
```
